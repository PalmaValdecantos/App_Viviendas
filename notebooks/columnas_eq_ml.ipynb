```python
# 1 de casas y otro de pisos
# Media precio distrito
# Desviación precio distrito
import pandas as pd
# Avoiding scientific notation with Pandas
pd.set_option('display.float_format', lambda x: '%.3f' % x)
```

### DATAFRAMES 

```python
# Full
df = pd.read_csv('Madrid_houses_updated_clean_new.csv')
# Map booleans
df['is_renewal_needed'] = df['is_renewal_needed'].map({False: 0, True: 1})
df['has_lift'] = df['has_lift'].map({False: 0, True: 1})
df['is_exterior'] = df['is_exterior'].map({False: 0, True: 1})
df['has_parking'] = df['has_parking'].map({False: 0, True: 1})
df.drop(['Unnamed: 0', '_c0', 'id'], axis=1, inplace=True)
df
```

# Pisos

```python
# Dataframe 1 - Pisos
pisos = [1,3,4,5]
pisos_df = df[df.house_type.isin(pisos)]
```

```python
# Mapping districts with mean prices
pisos_df['mean_price_district'] = pisos_df['district'].map(dict(pisos_df.groupby('district')['buy_price'].mean()))
pisos_df
```

### Mapping districts with std deviation in prices

```python
# Mapping districts with std deviation in prices
pisos_df['std_deviation_price_district'] = pisos_df['district'].map(dict(pisos_df.groupby('district')['buy_price'].std()))
pisos_df.head()
```

```python
# Dummies
pisos_dummies = pd.get_dummies(pisos_df['house_type'])
pisos_con_dummies = pd.concat([pisos_df, pisos_dummies], axis=1)
pisos_con_dummies.drop(['house_type'],axis=1, inplace=True)
pisos_con_dummies
```

----

```python
# DataFrame 2 - Chalets
chalets_df = df[~df['house_type'].isin(pisos)]
```

```python
# Mapping districts with mean prices
chalets_df['mean_price_district'] = chalets_df['district'].map(dict(chalets_df.groupby('district')['buy_price'].mean()))
chalets_df
```

```python
# Mapping districts with std deviation in prices
chalets_df['std_deviation_price_district'] = chalets_df['district'].map(dict(chalets_df.groupby('district')['buy_price'].std()))
chalets_df
```

## CSV


```python
pisos_con_dummies.to_csv('pisos_nuevo.csv', index=False)
chalets_df.to_csv('chalets_nuevo.csv', index=False)
```
